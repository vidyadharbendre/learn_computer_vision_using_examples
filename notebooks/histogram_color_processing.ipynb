```python
# Import required libraries
import cv2  # OpenCV library for image processing
import numpy as np  # NumPy for numerical operations
import matplotlib.pyplot as plt  # Matplotlib for plotting
import os  # os module for working with file paths

# Define the image path using os.path.join
# This ensures that the image path is correctly formed regardless of the operating system
image_path = os.path.join(os.getcwd(), 'data', 'images', 'worldview-1.jpg')  # Modify this with your image path

# Load the image using OpenCV
# cv2.imread loads the image in BGR format by default (Blue-Green-Red)
image = cv2.imread(image_path)

# Convert the image from BGR to RGB format
# OpenCV uses BGR, but most image processing libraries, like matplotlib, expect RGB format
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Calculate histograms for each color channel: Red (R), Green (G), Blue (B)
# cv2.calcHist calculates the histogram of an image
# It takes the image, the color channel index (0 for Red, 1 for Green, and 2 for Blue),
# an optional mask (None means no mask), number of bins (256 for pixel values between 0 and 255),
# and the range of pixel values (0 to 256 for each channel)
hist_r = cv2.calcHist([image_rgb], [0], None, [256], [0, 256])  # Histogram for Red channel
hist_g = cv2.calcHist([image_rgb], [1], None, [256], [0, 256])  # Histogram for Green channel
hist_b = cv2.calcHist([image_rgb], [2], None, [256], [0, 256])  # Histogram for Blue channel

# Plot the histograms for each color channel using matplotlib
# A histogram shows the frequency of pixel values (0-255) for each color channel
plt.figure(figsize=(10, 5))  # Set the figure size for the plot
plt.title("Color Histograms")  # Set the title of the plot

# Plot the histograms for each channel with appropriate colors (Red, Green, Blue)
plt.plot(hist_r, color='r', label='Red')  # Red channel histogram
plt.plot(hist_g, color='g', label='Green')  # Green channel histogram
plt.plot(hist_b, color='b', label='Blue')  # Blue channel histogram

# Set the x-axis limits to cover all pixel values (0-255)
plt.xlim(0, 256)

# Add a legend to identify each color channel
plt.legend()

# Display the plot
plt.show()
```
