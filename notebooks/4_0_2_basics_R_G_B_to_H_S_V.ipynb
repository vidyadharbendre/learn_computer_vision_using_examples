```python
"""
This script demonstrates how to convert an RGB image to the HSV color space using OpenCV 
and extract individual HSV channels (Hue, Saturation, and Value).

### Key Concepts:

1. **RGB to HSV Conversion:**
   - RGB (Red, Green, Blue): A widely used color model representing colors as intensities of red, green, and blue.
   - HSV (Hue, Saturation, Value): A color model that represents colors more intuitively:
     - **Hue (H):** Represents the type of color, ranging from 0° to 360° (encoded as 0-179 in OpenCV).
     - **Saturation (S):** Represents the intensity or purity of the color (0-255 in OpenCV).
     - **Value (V):** Represents the brightness of the strongest color in a pixel (0-255 in OpenCV).

2. **Purpose of HSV Color Space:**
   - Separates chromatic content (Hue and Saturation) from intensity (Value).
   - Useful for color-based image processing tasks, such as object detection, segmentation, and filtering.

3. **Program Workflow:**
   - **Input:** A predefined 3x3 RGB image represented as a NumPy array.
     - Colors include primary colors (Red, Green, Blue), secondary colors (Yellow, Cyan, Magenta), 
       and shades of gray.
   - **RGB to HSV Conversion:** The `cv2.cvtColor` function converts the image from the RGB to HSV color space.
   - **Channel Extraction:** Individual channels (H, S, V) are extracted using NumPy slicing:
     - **Hue (H):** First channel of the HSV image.
     - **Saturation (S):** Second channel of the HSV image.
     - **Value (V):** Third channel of the HSV image.
   - **Output:** Prints the pixel values of each HSV channel for analysis.

4. **Program Output Explanation:**
   - **Hue (H):** Displays the color types of the image pixels as encoded angular values.
   - **Saturation (S):** Shows the intensity/purity of each color pixel.
   - **Value (V):** Represents the brightness levels of the strongest color in each pixel.

5. **Use Cases:**
   - **Hue Analysis:** Identify specific colors in an image for segmentation or tracking.
   - **Saturation Analysis:** Evaluate color intensity for filtering faint colors or creating grayscale images.
   - **Value Analysis:** Analyze brightness levels for preprocessing tasks like thresholding or lighting adjustments.

6. **Program Applications:**
   - Color-based object detection and segmentation.
   - Feature extraction based on specific HSV properties.
   - Preprocessing images for machine learning and computer vision tasks.

"""

import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the 3x3 RGB image
image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],       # Row 1: Red, Green, Blue
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]], # Row 2: Yellow, Cyan, Magenta
    [[128, 128, 128], [64, 64, 64], [32, 32, 32]]  # Row 3: Shades of gray
], dtype=np.uint8)

# Convert the RGB image to HSV using OpenCV
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# Extract the channels
H_channel = hsv_image[:, :, 0]  # Hue
S_channel = hsv_image[:, :, 1]  # Saturation
V_channel = hsv_image[:, :, 2]  # Value

print("Hue Channel (H):")
print(H_channel)

print("\nSaturation Channel (S):")
print(S_channel)

print("\nValue Channel (V):")
print(V_channel)
```

```python
"""
This script demonstrates the conversion of an RGB image to the HSV color space
and the calculation of the Value (V) component in the HSV model using both OpenCV 
and a manual approach.

### Key Concepts:

1. **Color Spaces:**
   - RGB: A standard color model where colors are represented as combinations of 
     Red (R), Green (G), and Blue (B) intensities.
   - HSV: A cylindrical color space that represents colors closer to human perception:
     - Hue (H): Specifies the type of color (e.g., red, blue, green).
     - Saturation (S): Indicates the purity or vividness of the color.
     - Value (V): Represents the brightness or intensity of the most dominant color.

2. **Value (V) in HSV:**
   - Definition: The Value (V) component represents the brightness of the strongest color 
     in a pixel. 
   - Formula: 
       V = max(R, G, B)
   - It highlights the most prominent or "brightest" color in each pixel, making it useful
     for tasks like filtering and thresholding based on brightness.

3. **Program Workflow:**
   - A 3x3 RGB image is defined using a NumPy array, containing primary colors 
     (Red, Green, Blue), secondary colors (Yellow, Cyan, Magenta), and grayscale shades.
   - The RGB image is converted to the HSV color space using OpenCV's `cvtColor` function.
   - The Value (V) channel is extracted directly from the HSV image.
   - The Value (V) component is manually calculated by taking the maximum of the R, G, and B 
     values for each pixel.
   - The program compares the Value (V) obtained from OpenCV with the manually calculated 
     Value (V).

4. **Use Cases:**
   - Image processing tasks that rely on brightness (e.g., highlighting or filtering based on 
     lightness or intensity).
   - Identifying the most dominant color in a pixel for applications like segmentation or 
     color-based filtering.

5. **Comparison:**
   - The program outputs the Value (V) calculated by OpenCV and the manual approach to 
     demonstrate that they yield identical results.

"""

import numpy as np
import cv2

# Define the 3x3 RGB image
image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],       # Row 1: Red, Green, Blue
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]], # Row 2: Yellow, Cyan, Magenta
    [[128, 128, 128], [64, 64, 64], [32, 32, 32]]  # Row 3: Shades of gray
], dtype=np.uint8)

# Convert to HSV using OpenCV
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# Extract Value (V) channel from OpenCV
V_channel_opencv = hsv_image[:, :, 2]

# Manually calculate V (max of R, G, B)
V_channel_manual = np.max(image, axis=2)

# Compare results
print("Value (V) from OpenCV:")
print(V_channel_opencv)

print("\nManually Calculated Value (V):")
print(V_channel_manual)
```

```python
"""
This script demonstrates the concepts of color space conversion and manual calculations
of properties such as Value (V) in the HSV model and Intensity (I) in the HSI model.

### Key Concepts:

1. **Color Spaces:**
   - RGB: The standard color model where colors are represented using the combination of 
     Red (R), Green (G), and Blue (B) components.
   - HSV: A cylindrical color space with three components:
     - Hue (H): Represents the color type (e.g., red, green, blue).
     - Saturation (S): Indicates the vividness or purity of the color.
     - Value (V): Represents the brightness of the most dominant color in a pixel.
   - HSI: Another color space that focuses on how humans perceive colors:
     - Hue (H) and Saturation (S) are the same as in HSV.
     - Intensity (I): Represents the average lightness of a pixel.

2. **Conversions and Calculations:**
   - OpenCV's `cvtColor` function is used to convert the RGB image to the HSV color space.
   - The Value (V) component in HSV is the maximum of the R, G, and B values for each pixel.
   - The Intensity (I) component in HSI is the average of the R, G, and B values for each pixel.

3. **Comparison of V and I:**
   - Value (V): Sensitive to the brightest component in a pixel, highlighting the most dominant color.
   - Intensity (I): Represents the overall lightness, making it suitable for tasks involving grayscale processing.

### Program Workflow:
1. Define a 3x3 RGB image as a NumPy array, containing a variety of colors and grayscale shades.
2. Convert the RGB image to HSV using OpenCV and extract the Value (V) channel.
3. Manually calculate:
   - Value (V): Maximum of the R, G, and B values for each pixel.
   - Intensity (I): Average of the R, G, and B values for each pixel.
4. Print and compare the results to understand the differences between Value (V) and Intensity (I).

### Example Use Cases:
- Color-based filtering: Use HSV's Value (V) for brightness thresholds.
- Grayscale transformations: Use HSI's Intensity (I) to handle lightness.

"""

import numpy as np
import cv2

# Define the 3x3 RGB image
image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],       # Row 1: Red, Green, Blue
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]], # Row 2: Yellow, Cyan, Magenta
    [[128, 128, 128], [64, 64, 64], [32, 32, 32]]  # Row 3: Shades of gray
], dtype=np.uint8)

# Convert to HSV using OpenCV
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# Extract Value (V) channel from OpenCV
V_channel_opencv = hsv_image[:, :, 2]

# Manually calculate V (max of R, G, B)
V_channel_manual = np.max(image, axis=2)

# Manually calculate I (average of R, G, B)
I_channel_manual = np.mean(image, axis=2)

# Compare results
print("Value (V) from OpenCV:")
print(V_channel_opencv)

print("\nManually Calculated Value (V):")
print(V_channel_manual)

print("\nManually Calculated Intensity (I):")
print(I_channel_manual)
```
