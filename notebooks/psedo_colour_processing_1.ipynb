```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a grayscale image
# Specify the path to the image
image_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')  # Replace with your image path
#image_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')

gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load the image in grayscale mode

# List of colormaps to be applied for pseudo-color transformations
colormaps = [
    cv2.COLORMAP_JET,     # Colormap 1: Jet (popular for heatmap visualization)
    cv2.COLORMAP_VIRIDIS, # Colormap 2: Viridis (perceptually uniform colormap)
    cv2.COLORMAP_COOL     # Colormap 3: Cool (shades of cyan to magenta)
]

# Create a figure to display the images
plt.figure(figsize=(12, 8))  # Set the figure size (width x height in inches)

# Display the original grayscale image
plt.subplot(2, 2, 1)  # Create the first subplot (2 rows, 2 columns, position 1)
plt.imshow(gray_image, cmap='gray')  # Display the image in grayscale
plt.title("Original Grayscale")  # Title for the original grayscale image
plt.axis("off")  # Turn off axis labels for better visualization

# Apply and display pseudo-color transformations for each colormap
for i, colormap in enumerate(colormaps, 2):  # Start enumeration from 2 for subplot positions
    # Apply the current colormap to the grayscale image
    pseudo_color = cv2.applyColorMap(gray_image, colormap)
    
    # Display the pseudo-color image
    plt.subplot(2, 2, i)  # Create a subplot for each colormap (positions 2, 3, 4)
    plt.imshow(cv2.cvtColor(pseudo_color, cv2.COLOR_BGR2RGB))  # Convert BGR (OpenCV) to RGB (Matplotlib)
    plt.title(f"Colormap {colormap}")  # Title indicating the colormap used
    plt.axis("off")  # Turn off axis labels for better visualization

# Adjust layout to avoid overlapping titles and images
plt.tight_layout()

# Show all images in the figure
plt.show()
```
