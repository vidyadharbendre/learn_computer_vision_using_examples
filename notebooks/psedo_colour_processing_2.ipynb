```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a grayscale image
# Specify the path to the image
image_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')  # Replace with your image path
#image_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')

gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load the image in grayscale mode

# Create a list of selected colormaps to demonstrate pseudo-color transformations
colormaps = [
    cv2.COLORMAP_JET,     # Colormap 2
    cv2.COLORMAP_PARULA,  # Colormap 4
    cv2.COLORMAP_TURBO,   # Colormap 8
    cv2.COLORMAP_HOT      # Colormap 16
]

# Initialize the figure for displaying the images
plt.figure(figsize=(16, 12))

# Row 1: Original grayscale image and its default color representation
plt.subplot(3, 2, 1)  # First row, first column
plt.imshow(gray_image, cmap='gray')  # Display grayscale image
plt.title("Original Grayscale")
plt.axis("off")  # Remove axes for better visualization

plt.subplot(3, 2, 2)  # First row, second column
pseudo_color1 = cv2.applyColorMap(gray_image, colormaps[0])  # Apply Colormap 2
plt.imshow(cv2.cvtColor(pseudo_color1, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for Matplotlib
plt.title("Colormap 2 (JET)")
plt.axis("off")

# Row 2: Colormaps 4 and 8
plt.subplot(3, 2, 3)  # Second row, first column
pseudo_color2 = cv2.applyColorMap(gray_image, colormaps[1])  # Apply Colormap 4
plt.imshow(cv2.cvtColor(pseudo_color2, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title("Colormap 4 (PARULA)")
plt.axis("off")

plt.subplot(3, 2, 4)  # Second row, second column
pseudo_color3 = cv2.applyColorMap(gray_image, colormaps[2])  # Apply Colormap 8
plt.imshow(cv2.cvtColor(pseudo_color3, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title("Colormap 8 (TURBO)")
plt.axis("off")

# Row 3: Colormaps 16 and an alternate color representation
plt.subplot(3, 2, 5)  # Third row, first column
pseudo_color4 = cv2.applyColorMap(gray_image, colormaps[3])  # Apply Colormap 16
plt.imshow(cv2.cvtColor(pseudo_color4, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title("Colormap 16 (HOT)")
plt.axis("off")

# Add another interesting colormap if needed
plt.subplot(3, 2, 6)  # Third row, second column
pseudo_color5 = cv2.applyColorMap(gray_image, cv2.COLORMAP_COOL)  # Example alternative
plt.imshow(cv2.cvtColor(pseudo_color5, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title("Alternate Colormap (COOL)")
plt.axis("off")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()

```
