```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load images
roses_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')
lena_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')

# Load the original images in grayscale
#noisy_image = cv2.imread(roses_path, cv2.IMREAD_GRAYSCALE)
noisy_image = cv2.imread(lena_path, cv2.IMREAD_GRAYSCALE)

# Apply Gaussian filtering (for comparison)
gaussian_filtered = cv2.GaussianBlur(noisy_image, (5, 5), 0)

# Apply Median filtering to remove salt-and-pepper noise
median_filtered = cv2.medianBlur(noisy_image, 5)

# Plot the results for comparison
plt.figure(figsize=(20, 7))

# Original Noisy Image
plt.subplot(1, 3, 1)
plt.imshow(noisy_image, cmap='gray')
plt.title('Original Noisy Image')
plt.axis('off')

# Gaussian Filtered (for comparison)
plt.subplot(1, 3, 2)
plt.imshow(gaussian_filtered, cmap='gray')
plt.title('Gaussian Filtered')
plt.axis('off')

# Median Filtered (Noise Reduced)
plt.subplot(1, 3, 3)
plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filtered (Noise Reduced)')
plt.axis('off')

plt.show()
```

```python
# Explanation of Output
# =====================

# Original Noisy Image:
# This plot shows the initial image with strong salt-and-pepper noise. 
# Salt-and-pepper noise appears as random white and black pixels scattered across the image,
# making the image appear grainy and affecting the quality of details.

# Gaussian Filtered:
# The Gaussian filter smooths the image by averaging the pixel values in a neighborhood,
# which reduces noise but also causes blurring across the entire image, including the edges.
# As a result, important details like edges and fine structures are blurred,
# which may not be ideal for applications that rely on edge clarity.

# Median Filtered:
# The median filter is applied to specifically target and reduce salt-and-pepper noise.
# It replaces each pixel’s value with the median of its neighborhood, effectively removing 
# outlier pixels (the noise) while preserving edges and fine details better than Gaussian filtering.
# This method is especially effective for salt-and-pepper noise, as it selectively reduces noise 
# without excessively blurring the entire image.

# Overall Approach:
# The use of median filtering helps retain the original signal or structure in the image
# while effectively reducing the prominent noise, resulting in an image that’s 
# much cleaner and more suitable for further processing steps.
```

```python
# Explanation of Noise and Filtering Challenges
# =============================================

# Details on Noise and Filtering with Images:
# - Every Pixel Has Noise:
#   In low-light images, each pixel is affected by noise, often causing 
#   random variations in pixel values that obscure details and clarity.
#   This can result in a grainy or mottled appearance, especially visible in smooth areas.

# - Effect of Low Light:
#   Capturing an image in low light conditions increases the chance of noise 
#   across the entire image, as the camera sensor tries to compensate for the lack of light.
#   This type of noise can obscure finer details and make the image harder to process accurately.

# - Increasing Filter Size:
#   Applying filters with a larger kernel size helps reduce noise by averaging
#   pixel values over a broader area. This can improve the overall clarity of the image
#   by smoothing out noise; however, it also risks introducing unwanted texture or blurring
#   in smooth regions, especially in the background. This smoothing effect can impact 
#   finer details, potentially affecting the interpretation of important regions.

# - Limitations of Noise Removal:
#   Although larger filters can reduce noise effectively, they do not fully eliminate it.
#   Subtle noise artifacts may still be present, and increasing filter strength may
#   lead to a trade-off by diminishing important image details alongside the noise.
```
