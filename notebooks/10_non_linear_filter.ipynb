```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image (you can replace this with any image path)
image_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')

# Load the original image in grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Apply non-linear filter (median filter)
non_linear_filtered_image = cv2.medianBlur(image, 3)
```

```python
# plot the original image and its magnitude spectrum
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(non_linear_filtered_image, cmap='gray'), plt.title('Non Linear Filtered Image')
plt.show()
```
