```python
# Theory: Color Complements in Image Processing
#
# In color theory, a complement is a color that, when combined with a given color, produces a neutral color such as white or gray.
# In image processing, the complement of a color is derived by subtracting each color channel (RGB) value from 255.
#
# The complement of primary colors:
# - Red (255, 0, 0) becomes Cyan (0, 255, 255)
# - Green (0, 255, 0) becomes Magenta (255, 0, 255)
# - Blue (0, 0, 255) becomes Yellow (255, 255, 0)
#
# Complementary color effects are often used in photography, image manipulation, and art to create visual contrasts and highlights.
# For example, applying color complements in image processing can help in:
# 1. Enhancing details in dark regions.
# 2. Creating visually striking effects similar to photographic negatives.
# 3. Highlighting contrasts in color-dominant areas.
#
# This program demonstrates how to compute and display the complement of an image using Python and OpenCV.
# The image is loaded, and its complement is calculated by subtracting each pixel's RGB values from 255. Both the original 
# and complemented images are then displayed for comparison.

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def color_complement(image_path):
    # Load the image
    original = cv2.imread(image_path)
    original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    
    # Compute the complement
    complement = 255 - original  # Subtract pixel values from 255
    
    # Plot the results
    plt.figure(figsize=(10, 5))
    
    # Display original image
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title("Original Image")
    plt.axis('off')
    
    # Display complement image
    plt.subplot(1, 2, 2)
    plt.imshow(complement)
    plt.title("Complement Image")
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Set the path to the images directory and specify 'roses.jpg'
image_path = 'data/images/roses.jpg'

# Check if the 'roses.jpg' exists
if os.path.exists(image_path):
    color_complement(image_path)
else:
    print(f"The image 'roses.jpg' was not found in the '{image_directory}' directory.")
```
