```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
#image_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')  # Create the full path to the image file
image_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')

rgb_image = cv2.imread(image_path)  # Read the image using OpenCV (in BGR format by default)
rgb_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB)  # Convert the image from BGR to RGB format for proper color representation

# Separate the channels
R = rgb_image[:, :, 0]  # Extract the Red channel (all rows and columns, but only the first color channel)
G = rgb_image[:, :, 1]  # Extract the Green channel (all rows and columns, but only the second color channel)
B = rgb_image[:, :, 2]  # Extract the Blue channel (all rows and columns, but only the third color channel)

# Display the channels using Matplotlib
plt.figure(figsize=(12, 8))  # Set the figure size for better visualization

# Original RGB Image
plt.subplot(2, 2, 1)  # Create a subplot in a 2x2 grid at position 1
plt.imshow(rgb_image)  # Display the original RGB image
plt.title("Original RGB Image")  # Set the title for the subplot
plt.axis("off")  # Turn off axis labels for cleaner visualization

# Red Channel
plt.subplot(2, 2, 2)  # Create a subplot in a 2x2 grid at position 2
plt.imshow(R, cmap="Reds")  # Display the Red channel using a colormap that emphasizes red tones
plt.title("Red Channel")  # Set the title for the subplot
plt.axis("off")  # Turn off axis labels

# Green Channel
plt.subplot(2, 2, 3)  # Create a subplot in a 2x2 grid at position 3
plt.imshow(G, cmap="Greens")  # Display the Green channel using a colormap that emphasizes green tones
plt.title("Green Channel")  # Set the title for the subplot
plt.axis("off")  # Turn off axis labels

# Blue Channel
plt.subplot(2, 2, 4)  # Create a subplot in a 2x2 grid at position 4
plt.imshow(B, cmap="Blues")  # Display the Blue channel using a colormap that emphasizes blue tones
plt.title("Blue Channel")  # Set the title for the subplot
plt.axis("off")  # Turn off axis labels

# Adjust the layout and display the plots
plt.tight_layout()  # Adjust subplot spacing to minimize overlap
plt.show()  # Show the figure
```
