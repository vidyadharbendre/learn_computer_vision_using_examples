```python
"""
Pseudocolor Image Processing with Python

Pseudocolor image processing is a technique that combines multiple grayscale images 
captured in different spectral bands (e.g., Red, Green, Blue, Infrared) into a single 
RGB composite image. This enhances visualization by emphasizing specific features 
like vegetation, surface characteristics, or temperature variations.

Applications:
1. Remote sensing (e.g., vegetation mapping using multispectral satellite images).
2. Space exploration (e.g., analyzing planetary surfaces).
3. Multispectral imaging in medical and industrial fields.

Steps in this Program:
1. Load grayscale images corresponding to different spectral bands.
2. Combine these grayscale images into RGB composites using OpenCV.
3. Display the original grayscale images and their RGB composites using Matplotlib.

Dependencies:
- Python 3.x
- OpenCV for image processing
- NumPy for numerical operations
- Matplotlib for visualization
"""

# Composite 1: IR-G-B
# - Red Channel: Infrared (IR) band
# - Green Channel: Green (G) band
# - Blue Channel: Blue (B) band
composite_1 = cv2.merge([images['IR'], images['G'], images['B']])

# Composite 2: IR-R-B
# - Red Channel: Infrared (IR) band
# - Green Channel: Red (R) band
# - Blue Channel: Blue (B) band
composite_2 = cv2.merge([images['IR'], images['R'], images['B']])

# Composite 3: RRB Composite (Red-Red-Blue)
# - Red Channel: Red (R) band (repeated)
# - Green Channel: Red (R) band
# - Blue Channel: Blue (B) band
# This composite emphasizes features represented by Red and Blue channels.
rrb_composite = cv2.merge([images['R'], images['R'], images['B']])

# Step 4: Display images using Matplotlib
plt.figure(figsize=(15, 8))  # Define the figure size for the plot

# Display original grayscale images
# Each grayscale image corresponds to a spectral band (R, G, B, IR)
for i, (band, img) in enumerate(images.items(), start=1):
    plt.subplot(2, 4, i)  # Arrange plots in a 2x4 grid for better visualization
    plt.imshow(img, cmap='gray')  # Display the image in grayscale
    plt.title(f"{band} Band")    # Title for the image (e.g., "R Band")
    plt.axis('off')              # Remove axes for better visualization

# Display Composite 1: IR-G-B
# This composite maps the Infrared band to the Red channel.
plt.subplot(2, 4, 5)  # Position for Composite 1
plt.imshow(cv2.cvtColor(composite_1, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for Matplotlib
plt.title("Composite 1: IR-G-B")  # Title for Composite 1
plt.axis('off')                   # Remove axes

# Display Composite 2: IR-R-B
# This composite maps the Infrared band to the Red channel, Red band to Green, and Blue remains in Blue.
plt.subplot(2, 4, 6)  # Position for Composite 2
plt.imshow(cv2.cvtColor(composite_2, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for Matplotlib
plt.title("Composite 2: IR-R-B")  # Title for Composite 2
plt.axis('off')                   # Remove axes

# Display Composite 3: RRB
# This composite emphasizes Red and Blue bands.
plt.subplot(2, 4, 7)  # Position for Composite 3
plt.imshow(cv2.cvtColor(rrb_composite, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for Matplotlib
plt.title("Composite 3: RRB")  # Title for Composite 3
plt.axis('off')                   # Remove axes

# Adjust layout to prevent overlapping of plots and display the final visualization
plt.tight_layout()
plt.show()
```
