```python
import os
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

image_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')

# Load the image
image = cv2.imread(image_path)

# Convert to HLS color space (Hue, Lightness, Saturation)
hls_image = cv2.cvtColor(image, cv2.COLOR_BGR2HLS)

# Extract the lightness channel
lightness_channel = hls_image[:, :, 1]

# Flatten the array to create a 1D array of lightness values
lightness_values = lightness_channel.flatten()

# Create a DataFrame for easier plotting
df = pd.DataFrame(lightness_values, columns=['Lightness'])

# Create a box plot using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(y='Lightness', data=df, color='lightblue')
plt.title('Box Plot of Lightness Values')
plt.ylabel('Lightness Value')
plt.grid(axis='y')
plt.show()
```
