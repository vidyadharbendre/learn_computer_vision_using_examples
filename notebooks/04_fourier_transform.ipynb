```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Step 1: Load the image in grayscale
# Path to the image
image_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')
# Load the image
image = cv2.imread(image_path, 0)
```

```python
# Step 2: Perform Fourier Transform using numpy's fft2
f_transform = np.fft.fft2(image)
```

```python
# Step 3: Shift the zero frequency component (DC component) to the center
f_shift = np.fft.fftshift(f_transform)
```

```python
# Step 4: Calculate the magnitude spectrum
magnitude_spectrum = 20 * np.log(np.abs(f_shift))
```

```python
# Step 5: Plot the original image and its magnitude spectrum
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray'), plt.title('Magnitude Spectrum')
plt.show()
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Function to get the image path
def get_image_path(filename):
    return os.path.join(os.getcwd(), 'data', 'images', filename)

# Function to load the image in grayscale
def load_grayscale_image(image_path):
    grayscale_image = cv2.imread(image_path, 0)  # Load as grayscale
    if grayscale_image is None:
        raise FileNotFoundError(f"Error: Image could not be loaded. Please check the file path: {image_path}")
    return grayscale_image

# Function to apply Fourier Transform and return the transformed image and magnitude spectrum
def apply_fourier_transform(image):
    # Step 2: Perform Fourier Transform using numpy's fft2
    f_transform = np.fft.fft2(image)
    
    # Step 3: Shift the zero frequency component (DC component) to the center
    f_shift = np.fft.fftshift(f_transform)
    
    # Step 4: Calculate the magnitude spectrum
    magnitude_spectrum = 20 * np.log(np.abs(f_shift))
    
    return f_transform, magnitude_spectrum

# Function to display the original image and magnitude spectrum
def display_images(grayscale_image, magnitude_spectrum):
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    axs[0].imshow(grayscale_image, cmap='gray')
    axs[0].set_title('Original Image')
    axs[0].axis('off')

    axs[1].imshow(magnitude_spectrum, cmap='gray')
    axs[1].set_title('Magnitude Spectrum')
    axs[1].axis('off')

    plt.tight_layout()
    plt.show()

# Main function to handle everything
def main():
    # Get image path
    image_path = get_image_path('roses.jpg')
    print(f"Loading image from: {image_path}")
    
    # Load the image in grayscale
    grayscale_image = load_grayscale_image(image_path)
    
    # Apply Fourier Transform to the grayscale image
    f_transform, magnitude_spectrum = apply_fourier_transform(grayscale_image)
    
    # Display the original image and its magnitude spectrum
    display_images(grayscale_image, magnitude_spectrum)

# Execute the main function
if __name__ == "__main__":
    main()
```
