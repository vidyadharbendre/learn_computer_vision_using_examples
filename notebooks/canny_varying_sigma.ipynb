```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Load the image in grayscale
lena_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')
image = cv2.imread(lena_path, cv2.IMREAD_GRAYSCALE)

# Define a list of sigma values to test
sigma_values = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0]

# Set up subplots with 3 rows and 2 columns
plt.figure(figsize=(12, 12))

for i, sigma in enumerate(sigma_values):
    # Apply Gaussian Blur with varying sigma values
    blurred_image = cv2.GaussianBlur(image, (5, 5), sigma)
    
    # Apply the Canny edge detector
    low_threshold = 50
    high_threshold = 150
    edges = cv2.Canny(blurred_image, low_threshold, high_threshold)
    
    # Plot the edges detected for the current sigma in a 3x2 grid
    plt.subplot(3, 2, i + 1)
    plt.imshow(edges, cmap='gray')
    plt.title(f"Canny with Sigma = {sigma}")
    plt.axis("off")

# Display all images
plt.tight_layout()
plt.show()
```
