```python
"""
# Program for Enhancing Individual Color Channels in an Image

## Theoretical Background:
Color enhancement is a critical aspect of image processing, especially for applications such as computer vision, photography, and image editing. This program demonstrates how to selectively enhance individual color channels (Red, Green, and Blue) in an RGB image using OpenCV and NumPy.

### Key Concepts:
1. **RGB Color Model**:
   - An image in the RGB color model consists of three channels: Red, Green, and Blue. 
   - Each channel represents the intensity of the respective color at every pixel, with values typically ranging from 0 to 255.

2. **Enhancing Color Channels**:
   - Amplifying a specific color channel increases the intensity of that color in the image, making it more prominent.
   - This can be achieved by multiplying the pixel values of the desired channel by a factor greater than 1.

3. **Image Representation**:
   - Images are represented as multidimensional arrays (matrices) in OpenCV and NumPy. 
   - For an RGB image, the shape of the array is `(height, width, 3)`, where the last dimension corresponds to the three color channels.

4. **Matrix Operations**:
   - The program uses NumPy for element-wise operations on the image array to modify the color intensities.
   - The `np.clip()` function ensures that the pixel values remain within the valid range (0 to 255) after enhancement.

5. **Visualization**:
   - The `matplotlib` library is used to display the original and enhanced images side by side for comparison.

### Program Workflow:
1. Load an RGB image using OpenCV.
2. Convert the image from BGR (default in OpenCV) to RGB for correct color representation.
3. Create copies of the image to enhance individual channels (Red, Green, Blue):
   - Multiply the pixel values of the target channel by 1.5 to amplify the color intensity.
   - Use `np.clip()` to avoid overflow (values exceeding 255).
4. Display the original image along with the enhanced versions using Matplotlib.

### Applications:
- Enhancing visual appeal in photography.
- Highlighting specific features in medical imaging (e.g., red for blood vessels).
- Improving contrast and visibility in surveillance systems.
- Preprocessing for machine learning models requiring distinct color features.
"""

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Define the image path
image_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')  # Replace with your image path

# Load the RGB image
rgb_image = cv2.imread(image_path)

# Convert from BGR to RGB for display purposes
rgb_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB)

# Enhance red channel
red_emphasis = rgb_image.copy()
red_emphasis[:, :, 0] = np.clip(red_emphasis[:, :, 0] * 1.5, 0, 255)  # Amplify red by 50%

# Enhance green channel
green_emphasis = rgb_image.copy()
green_emphasis[:, :, 1] = np.clip(green_emphasis[:, :, 1] * 1.5, 0, 255)  # Amplify green by 50%

# Enhance blue channel
blue_emphasis = rgb_image.copy()
blue_emphasis[:, :, 2] = np.clip(blue_emphasis[:, :, 2] * 1.5, 0, 255)  # Amplify blue by 50%

# Visualize original and enhanced images
titles = ["Original Image", "Red Emphasis", "Green Emphasis", "Blue Emphasis"]
images = [rgb_image, red_emphasis, green_emphasis, blue_emphasis]

# Plot images
fig, axes = plt.subplots(1, 4, figsize=(20, 6))

for ax, title, img in zip(axes, titles, images):
    ax.imshow(img.astype(np.uint8))  # Ensure the image is in uint8 format
    ax.set_title(title)
    ax.axis('off')  # Hide axes

plt.tight_layout()
plt.show()
```
