```python
"""
# RGB Color Demonstration using Python

## Theoretical Concept:
RGB stands for Red, Green, and Blue, the three primary colors of light. 
In digital image processing, colors are represented as combinations of these three colors, 
where each color has an intensity value ranging from 0 to 255.

- **Red (255, 0, 0)**: Represents pure red color.
- **Green (0, 255, 0)**: Represents pure green color.
- **Blue (0, 0, 255)**: Represents pure blue color.
- **Other Colors**: Created by mixing red, green, and blue in various intensities.
    - Yellow (255, 255, 0): A mix of red and green.
    - Cyan (0, 255, 255): A mix of green and blue.
    - Magenta (255, 0, 255): A mix of red and blue.
    - White (255, 255, 255): All colors combined at full intensity.
    - Black (0, 0, 0): Absence of color or light.

This program demonstrates RGB colors using Python's matplotlib library by:
1. Defining a set of RGB colors.
2. Visualizing each color as a patch in a grid.
3. Displaying the name of each color as a title above its corresponding patch.

## Code Implementation:
"""

import matplotlib.pyplot as plt

# Define RGB colors
colors = {
    "Red": (255, 0, 0),
    "Green": (0, 255, 0),
    "Blue": (0, 0, 255),
    "Yellow": (255, 255, 0),
    "Cyan": (0, 255, 255),
    "Magenta": (255, 0, 255),
    "White": (255, 255, 255),
    "Black": (0, 0, 0),
}

# Create a figure to display the colors
plt.figure(figsize=(8, 6))
for i, (color_name, rgb) in enumerate(colors.items()):
    plt.subplot(2, 4, i + 1)
    plt.imshow([[rgb]])  # Display the color
    plt.axis('off')  # Hide axes
    plt.title(color_name)

plt.tight_layout()
plt.show()
```
