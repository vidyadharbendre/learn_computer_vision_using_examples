```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Function to get the image path
def get_image_path(filename):
    return os.path.join(os.getcwd(), 'data', 'images', filename)

# Function to load the image in both color and grayscale
def load_image(image_path):
    color_image = cv2.imread(image_path)
    
    # Convert to HLS color space (Hue, Lightness, Saturation)
    hls_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2HLS)
    
    if color_image is None or hls_image is None:
        raise FileNotFoundError(f"Error: Image could not be loaded. Please check the file path: {image_path}")
    return color_image, hls_image
```

```python
# Main function to handle everything
def main():
    image_path = get_image_path('roses.jpg')
    print(f"Loading image from: {image_path}")
    
    color_image, hls_image = load_image(image_path=image_path)

    # Extract the lightness channel
    lightness_channel = hls_image[:, :, 1]

    # Create a histogram of lightness values
    plt.hist(lightness_channel.ravel(), bins=256, range=[0,256], color='lightblue')
    plt.title('Histogram of Lightness Values')
    plt.xlabel('Lightness Value')
    plt.ylabel('Frequency')
    plt.xlim([0, 256])
    plt.show()
```

```python
# Execute the main function
if __name__ == "__main__":
    main()
```
