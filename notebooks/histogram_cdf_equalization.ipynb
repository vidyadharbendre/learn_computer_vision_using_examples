```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Define the image path (Replace with your own image path)
image_path = os.path.join(os.getcwd(), 'data', 'images', 'worldview-1.jpg')  # Modify this with your image path

# Load the image in grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded properly
if image is None:
    raise FileNotFoundError(f"Image not found at {image_path}")

# Calculate the histogram of the original grayscale image
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Calculate the histogram of the equalized image
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# Calculate the cumulative distribution function (CDF) for both histograms
cdf_original = hist_original.cumsum()
cdf_equalized = hist_equalized.cumsum()

# Normalize CDFs for plotting
cdf_original_normalized = (cdf_original - cdf_original.min()) * 255 / (cdf_original.max() - cdf_original.min())
cdf_equalized_normalized = (cdf_equalized - cdf_equalized.min()) * 255 / (cdf_equalized.max() - cdf_equalized.min())

# Plot the original and equalized images along with their histograms and CDFs
plt.figure(figsize=(12, 10))

# Plot original image and its histogram
plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.plot(hist_original, color='black')
plt.title("Histogram of Original Image")
plt.xlim(0, 256)

# Annotate histogram values on the plot
for i in range(0, 256, 32):  # Annotating every 32nd value
    plt.text(i, hist_original[i].item(), f'{int(hist_original[i].item())}', color='red', fontsize=8, verticalalignment='bottom')

# Plot original image CDF
plt.subplot(2, 3, 3)
plt.plot(cdf_original_normalized, color='blue')
plt.title("CDF of Original Image")
plt.xlim(0, 256)

# Annotate CDF values on the plot
for i in range(0, 256, 32):  # Annotating every 32nd value
    plt.text(i, cdf_original_normalized[i].item(), f'{int(cdf_original_normalized[i].item())}', color='green', fontsize=8, verticalalignment='bottom')

# Plot equalized image and its histogram
plt.subplot(2, 3, 4)
plt.imshow(equalized_image, cmap='gray')
plt.title("Equalized Grayscale Image")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.plot(hist_equalized, color='black')
plt.title("Histogram of Equalized Image")
plt.xlim(0, 256)

# Annotate histogram values on the plot
for i in range(0, 256, 32):  # Annotating every 32nd value
    plt.text(i, hist_equalized[i].item(), f'{int(hist_equalized[i].item())}', color='red', fontsize=8, verticalalignment='bottom')

# Plot equalized image CDF
plt.subplot(2, 3, 6)
plt.plot(cdf_equalized_normalized, color='red')
plt.title("CDF of Equalized Image")
plt.xlim(0, 256)

# Annotate CDF values on the plot
for i in range(0, 256, 32):  # Annotating every 32nd value
    plt.text(i, cdf_equalized_normalized[i].item(), f'{int(cdf_equalized_normalized[i].item())}', color='green', fontsize=8, verticalalignment='bottom')

# Show the plots
plt.tight_layout()
plt.show()

# Print histogram and CDF values for both images
print("Histogram Values for Original Image:")
for i in range(0, 256, 32):  # Print every 32nd histogram value
    print(f"Intensity {i}: {int(hist_original[i].item())}")

print("\nCDF Values for Original Image:")
for i in range(0, 256, 32):  # Print every 32nd CDF value
    print(f"Intensity {i}: {int(cdf_original[i].item())}")

print("\nHistogram Values for Equalized Image:")
for i in range(0, 256, 32):  # Print every 32nd histogram value
    print(f"Intensity {i}: {int(hist_equalized[i].item())}")

print("\nCDF Values for Equalized Image:")
for i in range(0, 256, 32):  # Print every 32nd CDF value
    print(f"Intensity {i}: {int(cdf_equalized[i].item())}")
```
