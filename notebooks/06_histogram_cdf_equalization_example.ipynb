```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a simple grayscale image as a 10x10 numpy array
image = np.array([
    [100, 102, 103,  99, 100, 101, 255, 255, 255, 255],
    [ 99, 100, 105, 104, 103, 100, 102, 255, 255, 255],
    [100,  99,  98, 102, 100, 101, 104, 255, 255, 255],
    [102, 101, 104,  99, 100, 105, 102, 255, 255, 255],
    [ 98, 100, 101, 104, 102, 100, 105,  98, 255, 255],
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255]
], dtype=np.uint8)

# Display the original image
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.show()
```

```python
# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the equalized image
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.show()
```

```python
# Calculate histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Plot the histogram
plt.plot(hist, color='r')
plt.title('Histogram of Original Image')
plt.show()

# Calculate CDF and plot
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

plt.plot(cdf_normalized, color='b')
plt.title('CDF of Original Image')
plt.show()

# Display the equalized image's histogram
hist_eq = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
plt.plot(hist_eq, color='r')
plt.title('Histogram of Equalized Image')
plt.show()
```
