```python
"""
# Program Explanation

## Purpose:
This program demonstrates how to process an RGB image as a 3D NumPy array and extract its individual Red, Green, and Blue channels. Each pixel in an RGB image is represented by three values, corresponding to the intensities of the Red, Green, and Blue components.

## Key Concepts:
1. **RGB Image Representation:**
   - A color image is represented as a 3D matrix (NumPy array) of dimensions `Height x Width x 3`.
   - The three channels are:
     - Red channel: Contains the intensity values for the red component of each pixel.
     - Green channel: Contains the intensity values for the green component.
     - Blue channel: Contains the intensity values for the blue component.

2. **Image Matrix:**
   - The image is a 3x3 RGB image defined manually in the program.
   - Each pixel is represented as a list `[R, G, B]`:
     - Example: `[255, 0, 0]` represents a fully red pixel.

3. **Channel Extraction:**
   - To extract individual channels (Red, Green, and Blue), slicing is performed on the third dimension (index 0, 1, or 2) of the 3D array:
     - `image[:, :, 0]` extracts the Red channel.
     - `image[:, :, 1]` extracts the Green channel.
     - `image[:, :, 2]` extracts the Blue channel.

4. **Matrix Representation:**
   - After slicing, the Red, Green, and Blue channels are represented as 2D matrices of the same size as the original image (Height x Width).
   - Each value in the matrix corresponds to the intensity of the respective color channel for that pixel.

5. **Console Output:**
   - The program prints the extracted Red, Green, and Blue channels as 2D matrices, helping to visualize how each color channel contributes to the original image.

## Hypothetical Example Used:
The example image is a 3x3 RGB matrix with specific color values:
- First row contains pure Red, Green, and Blue pixels.
- Second row contains Yellow, Cyan, and Magenta (combinations of two primary colors).
- Third row contains shades of gray (equal intensity of R, G, and B).

## Practical Usage:
This technique forms the foundation for many image processing tasks, such as:
- Modifying color channels separately.
- Analyzing specific color components.
- Converting RGB images to other color spaces (e.g., HSI, grayscale).

## Output:
- Red Channel: Displays the intensity values of the red component.
- Green Channel: Displays the intensity values of the green component.
- Blue Channel: Displays the intensity values of the blue component.

"""

import os
import numpy as np
import matplotlib.pyplot as plt
import cv2


# Example: A simple 3x3 pixel RGB image (values are hypothetical)
image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],  # Row 1: Red, Green, Blue
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],  # Row 2: Yellow, Cyan, Magenta
    [[128, 128, 128], [64, 64, 64], [32, 32, 32]]   # Row 3: Shades of gray
])

# Separate the channels manually using slicing
R_channel = image[:, :, 0]  # Red channel
G_channel = image[:, :, 1]  # Green channel
B_channel = image[:, :, 2]  # Blue channel

# Print each channel to demonstrate the process
print("Red Channel:")
print(R_channel)

print("\nGreen Channel:")
print(G_channel)

print("\nBlue Channel:")
print(B_channel)
```

```python
"""
# Program Explanation

## Purpose:
This program demonstrates how to manually extract the Red, Green, and Blue (RGB) channels from a simple 3x3 RGB image represented as a nested list. The process avoids using advanced libraries like NumPy and provides a basic understanding of RGB channel separation.

## Key Concepts:
1. **RGB Image Representation:**
   - An RGB image is represented as a list of lists, where each inner list corresponds to a row of pixels.
   - Each pixel is a list of three values `[R, G, B]`, representing the intensities of the Red, Green, and Blue channels, respectively.

2. **Manual Channel Extraction:**
   - To extract individual channels (Red, Green, and Blue), list comprehensions are used to traverse each pixel in the image:
     - `pixel[0]`: Extracts the Red component.
     - `pixel[1]`: Extracts the Green component.
     - `pixel[2]`: Extracts the Blue component.
   - The resulting channels are stored as 2D lists, where each value corresponds to the intensity of the respective color channel for that pixel.

3. **Image Structure in the Example:**
   - The image is a simple 3x3 grid:
     - Row 1: Pure colors (Red, Green, Blue).
     - Row 2: Combinations of two primary colors (Yellow, Cyan, Magenta).
     - Row 3: Shades of gray (equal intensities of R, G, and B).
   - This structure helps to clearly observe how each channel contributes to the image's colors.

4. **Channel Representation:**
   - The extracted Red, Green, and Blue channels are represented as 2D lists, making it easier to visualize and manipulate individual color components.

5. **Console Output:**
   - The program prints the extracted channels:
     - Red Channel: Displays only the red intensities for all pixels.
     - Green Channel: Displays only the green intensities for all pixels.
     - Blue Channel: Displays only the blue intensities for all pixels.

6. **Practical Application:**
   - Understanding channel separation is fundamental to color image processing.
   - It allows for channel-specific modifications (e.g., enhancing brightness in a specific channel or converting RGB to grayscale).

## Hypothetical Example Used:
The image is defined as:
- Row 1: `[[255, 0, 0], [0, 255, 0], [0, 0, 255]]` (Red, Green, Blue).
- Row 2: `[[255, 255, 0], [0, 255, 255], [255, 0, 255]]` (Yellow, Cyan, Magenta).
- Row 3: `[[128, 128, 128], [64, 64, 64], [32, 32, 32]]` (Shades of gray).

## Output:
- Red Channel: A 2D list of red intensities.
- Green Channel: A 2D list of green intensities.
- Blue Channel: A 2D list of blue intensities.

"""

# Example: A simple 3x3 RGB image
image = [
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],  # Row 1
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],  # Row 2
    [[128, 128, 128], [64, 64, 64], [32, 32, 32]]  # Row 3
]

# Initialize empty channels
R_channel = [[pixel[0] for pixel in row] for row in image]
G_channel = [[pixel[1] for pixel in row] for row in image]
B_channel = [[pixel[2] for pixel in row] for row in image]

print("Red Channel:", R_channel)
print("Green Channel:", G_channel)
print("Blue Channel:", B_channel)
```

```python
# Example: A simple 3x3 RGB image
image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],  # Row 1: Red, Green, Blue
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],  # Row 2: Yellow, Cyan, Magenta
    [[128, 128, 128], [64, 64, 64], [32, 32, 32]]   # Row 3: Shades of gray
])

# Extract the R, G, and B channels
R_channel = image[:, :, 0]  # Red
G_channel = image[:, :, 1]  # Green
B_channel = image[:, :, 2]  # Blue

# Create an empty canvas for displaying channels
R_display = np.zeros_like(image)
G_display = np.zeros_like(image)
B_display = np.zeros_like(image)

# Assign values for display
R_display[:, :, 0] = R_channel  # Assign Red channel to Red layer
G_display[:, :, 1] = G_channel  # Assign Green channel to Green layer
B_display[:, :, 2] = B_channel  # Assign Blue channel to Blue layer

# Plot the images
plt.figure(figsize=(10, 5))

# Original RGB image
plt.subplot(1, 4, 1)
plt.imshow(image)
plt.title("Original RGB Image")
plt.axis('off')

# Red channel
plt.subplot(1, 4, 2)
plt.imshow(R_display)
plt.title("Red Channel")
plt.axis('off')

# Green channel
plt.subplot(1, 4, 3)
plt.imshow(G_display)
plt.title("Green Channel")
plt.axis('off')

# Blue channel
plt.subplot(1, 4, 4)
plt.imshow(B_display)
plt.title("Blue Channel")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Step 1: Load the image using OpenCV
image_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')  # Modify with your image path
#image_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')  # Modify this with your image path
#image_path = os.path.join(os.getcwd(), 'data', 'images', 'vegetation_image_1.jpeg')  # Modify this with your image path


image = cv2.imread(image_path)  # Load as BGR format

# Step 2: Convert BGR to RGB (OpenCV loads images in BGR format by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 3: Split the RGB channels
R_channel = image_rgb[:, :, 0]  # Red channel
G_channel = image_rgb[:, :, 1]  # Green channel
B_channel = image_rgb[:, :, 2]  # Blue channel

# Step 4: Print information about the image
print(f"Image Dimensions: {image_rgb.shape}")
print(f"Red Channel: {R_channel.shape}")
print(f"Green Channel: {G_channel.shape}")
print(f"Blue Channel: {B_channel.shape}")

# Step 5: Access and print specific pixel values
x, y = 100, 150  # Example pixel coordinates
red_value = R_channel[y, x]
green_value = G_channel[y, x]
blue_value = B_channel[y, x]
print(f"Pixel at ({x}, {y}): R={red_value}, G={green_value}, B={blue_value}")

# Step 6: Visualize the original image and its RGB channels using Matplotlib
plt.figure(figsize=(12, 6))

# Original image
plt.subplot(1, 4, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

# Red channel
plt.subplot(1, 4, 2)
plt.imshow(R_channel, cmap='Reds')
plt.title("Red Channel")
plt.axis("off")

# Green channel
plt.subplot(1, 4, 3)
plt.imshow(G_channel, cmap='Greens')
plt.title("Green Channel")
plt.axis("off")

# Blue channel
plt.subplot(1, 4, 4)
plt.imshow(B_channel, cmap='Blues')
plt.title("Blue Channel")
plt.axis("off")

plt.tight_layout()
plt.show()
```
