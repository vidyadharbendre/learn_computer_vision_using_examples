```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Load the image
lena_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')
image = cv2.imread(lena_path, cv2.IMREAD_GRAYSCALE)

# Step 1: Apply Gaussian Blur with 5x5 kernel
blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)

# Step 2: Canny Edge Detection with custom thresholds
low_threshold = 50  # Low threshold for weak edges
high_threshold = 150  # High threshold for strong edges

edges = cv2.Canny(blurred_image, low_threshold, high_threshold)

# Display the results
plt.figure(figsize=(12, 8))

# Original image
plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Blurred image with 5x5 Gaussian kernel
plt.subplot(2, 3, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title("Gaussian Blur (5x5)")
plt.axis("off")

# Canny Edge Detection result
plt.subplot(2, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title(f"Canny Edges\nThresholds: {low_threshold}-{high_threshold}")
plt.axis("off")

plt.tight_layout()
plt.show()
```
