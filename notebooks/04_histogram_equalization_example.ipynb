```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt



image_path = os.path.join(os.getcwd(), 'data', 'images', 'roses.jpg')

# Load the original color image
image = cv2.imread(image_path)



image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for displaying

# Convert to YUV color space (Y = luminance, U and V = chrominance)
image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

# Apply histogram equalization on the Y channel (luminance)
image_yuv[:, :, 0] = cv2.equalizeHist(image_yuv[:, :, 0])

# Convert the equalized YUV image back to RGB format
image_equalized = cv2.cvtColor(image_yuv, cv2.COLOR_YUV2RGB)

# Plot the original and equalized images side by side
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Equalized Image
plt.subplot(1, 2, 2)
plt.imshow(image_equalized)
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```
