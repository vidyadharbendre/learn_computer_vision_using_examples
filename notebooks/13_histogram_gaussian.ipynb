```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 0        # Mean
sigma = 1     # Standard deviation
num_samples = 1000

# Generate Gaussian noise
gaussian_noise = np.random.normal(mu, sigma, num_samples)

# Plotting
plt.hist(gaussian_noise, bins=30, density=True)
plt.title('Histogram of Gaussian Noise')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
# The Probability Density Function (PDF) is fundamental to understanding Gaussian noise, 
# as it describes the distribution of the noise values generated. Let’s break down the relationship between 
# the PDF and the Gaussian noise:

# Relationship Between PDF and Gaussian Noise

# What is a Probability Density Function (PDF)?
# The PDF is a statistical function that describes the likelihood of a continuous random variable 
# to take on a given value. For Gaussian noise, the PDF is represented by the bell-shaped curve of the normal distribution.

# Gaussian PDF Formula
# The PDF for a Gaussian distribution with mean (𝜇) and standard deviation (𝜎) is given by the formula:
# 
# f(x) = (1 / (σ * sqrt(2π))) * e^(-(x - μ)² / (2σ²))
#
# Here:
#  - x is the variable (the noise value).
#  - μ is the mean (average) of the distribution.
#  - σ is the standard deviation (spread) of the distribution.
#  - e is the base of the natural logarithm.

# Using PDF to Generate Gaussian Noise
# When you generate Gaussian noise using NumPy’s np.random.normal(mu, sigma, num_samples), 
# you're effectively sampling from the normal distribution defined by the PDF. Each generated noise value 
# is a realization of the underlying Gaussian distribution characterized by the specified mean and standard deviation.

# Visualizing the PDF
# You can plot the PDF of Gaussian noise alongside the histogram of the generated noise values 
# to see how well the generated samples fit the theoretical distribution. 
# Here’s an example of how to do this:
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats  # For statistical functions

# Parameters
mu = 0        # Mean
sigma = 1     # Standard deviation
num_samples = 1000  # Number of samples

# Generate Gaussian noise
gaussian_noise = np.random.normal(mu, sigma, num_samples)

# Plotting
plt.figure(figsize=(10, 6))

# Histogram of generated noise
plt.hist(gaussian_noise, bins=30, density=True, alpha=0.6, color='gray', label='Histogram of Noise')

# Generate x values for the PDF
x = np.linspace(-4, 4, 1000)

# Calculate the PDF using the Gaussian formula
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Plot the PDF
plt.plot(x, pdf, 'r-', label='PDF of Gaussian Distribution')

# Labels and title
plt.title('Histogram of Gaussian Noise and its PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```
