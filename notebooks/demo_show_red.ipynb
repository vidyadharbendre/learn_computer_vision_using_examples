```python
import numpy as np
import matplotlib.pyplot as plt

# Create a uniform grayscale image (all values 255)
R = np.full((100, 100), 255, dtype=np.uint8)  # 100x100 image with all values 255

# Display using a grayscale colormap
plt.subplot(1, 2, 1)
plt.imshow(R, cmap='gray')  # Grayscale representation
plt.title('Grayscale (White)')
plt.axis('off')

# Display using the 'Reds' colormap
plt.subplot(1, 2, 2)
plt.imshow(R, cmap='Reds')  # Mapped to the red spectrum
plt.title('Reds Colormap (Bright Red)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
