```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Path to the image
image_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')

# Load the image in grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Error: Image could not be loaded.")
else:
    # Apply Gaussian Blur to smooth the image
    blurred_image = cv2.GaussianBlur(image, (5, 5), sigmaX=1.4)

    # Compute gradient in x direction
    gradient_x = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=3)

    # Compute gradient in y direction
    gradient_y = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=3)

    # Compute gradient magnitude (for visualization purposes)
    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
    gradient_magnitude = np.uint8(np.clip(gradient_magnitude, 0, 255))

    # Plot the results
    plt.figure(figsize=(20, 10))
    plt.subplot(1, 4, 1)
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 4, 2)
    plt.imshow(blurred_image, cmap='gray')
    plt.title("Gaussian Blurred Image")
    plt.axis("off")

    plt.subplot(1, 4, 3)
    plt.imshow(np.abs(gradient_x), cmap='gray')
    plt.title("Gradient in X Direction")
    plt.axis("off")

    plt.subplot(1, 4, 4)
    plt.imshow(np.abs(gradient_y), cmap='gray')
    plt.title("Gradient in Y Direction")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    # Plot gradient magnitude separately
    plt.figure(figsize=(7, 7))
    plt.imshow(gradient_magnitude, cmap='gray')
    plt.title("Gradient Magnitude")
    plt.axis("off")
    plt.show()
```
