```python
"""
# Intensity Slicing in Image Processing

## Overview
Intensity slicing, also known as level slicing, is a technique in image processing used to highlight specific 
ranges of intensity values in a grayscale image. 
It is particularly useful for feature detection and visualization when the image contains regions of interest 
with varying intensity levels.

## How It Works
1. **Define Intensity Ranges**: The grayscale intensity values (0-255) are divided into ranges (e.g., low, medium, high).
2. **Map to New Values**: Each intensity range is assigned a specific output intensity or color for better visualization.
3. **Enhance Interpretation**: Regions within the defined intensity ranges are visually distinct, making it easier to analyze the image.

## Applications
1. **Medical Imaging**: Highlighting specific tissue densities or anomalies in scans.
2. **Remote Sensing**: Identifying terrain features such as vegetation or water bodies.
3. **Feature Detection**: Enhancing regions of interest for further analysis.

## Program Overview
1. **Input**: A grayscale image.
2. **Define Ranges**: Intensity thresholds are specified for low, medium, and high ranges.
3. **Processing**: 
    - Pixels in the low-intensity range are assigned a specific value.
    - Pixels in the medium and high ranges are similarly mapped to new values.
4. **Output**: A modified image with visually distinct intensity-based regions.

## Key Benefits
- Enhances visualization by increasing the contrast between regions of interest.
- Simplifies analysis by grouping similar intensity values.
- Provides a quick way to segment and interpret grayscale images.

This program demonstrates intensity slicing using:
- **NumPy** for efficient array manipulation.
- **Matplotlib** for visualization.
"""

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load a grayscale image
# Use a sample grayscale image or create one using numpy
#image = np.random.randint(0, 256, (100, 100), dtype=np.uint8)  # Random grayscale image for demonstration

image_path = os.path.join(os.getcwd(), 'data', 'images', 'worldview-1.jpg')  # Replace with your image path
#image_path = os.path.join(os.getcwd(), 'data', 'images', 'lena_color.jpg')

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load the image in grayscale mode


# Step 2: Define intensity ranges
low, medium, high = 85, 170, 255  # Intensity thresholds

# Step 3: Apply intensity slicing
sliced_image = np.zeros_like(image)

# Assign intensity ranges
sliced_image[image <= low] = 50        # Low intensity - mapped to 50
sliced_image[(image > low) & (image <= medium)] = 150  # Medium intensity - mapped to 150
sliced_image[image > medium] = 250     # High intensity - mapped to 250

# Step 4: Visualize the results
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image, cmap="gray")
ax[0].set_title("Original Grayscale Image")
ax[0].axis("off")

ax[1].imshow(sliced_image, cmap="gray")
ax[1].set_title("Intensity Sliced Image")
ax[1].axis("off")

plt.tight_layout()
plt.show()
```
